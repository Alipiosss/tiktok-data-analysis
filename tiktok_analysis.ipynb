```python
# Create a countplot to visualize the count of each combination of claim_status and verified_status
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='claim_status', hue='verified_status')

# Add titles and labels
plt.title('Claim Status by Verification Status')
plt.xlabel('Claim Status')
plt.ylabel('Number of Videos')
plt.legend(title='Verified Status')
plt.tight_layout()
plt.show()
```

However, **verified users are slightly more likely to post opinion videos** than claim videos when compared proportionally. This could suggest that verified users, who may be more cautious about spreading unverified information, prefer to share content labeled as opinion rather than claim.


```python
# Set color palette manually: green for 'active', yellow for 'under review', orange for 'banned'
custom_palette = {
    'active': 'green',
    'under review': 'gold',
    'banned': 'orange'
}

# Create countplot to show the number of claim/opinion videos by author ban status
plt.figure(figsize=(6, 4))
sns.countplot(
    data=data,
    x='claim_status',
    hue='author_ban_status',
    palette=custom_palette,
    order=['claim', 'opinion'],
    hue_order=['active', 'under review', 'banned']
)

# Add labels and title
plt.title('Claim Status by Author Ban Status')
plt.xlabel('Claim Status')
plt.ylabel('Number of Videos')

# Add legend title
plt.legend(title='Author Ban Status')

# Improve layout
plt.tight_layout()
plt.show()

```

The bar chart shows that:

- **Active authors** are the most common across both claims and opinions. They represent the majority of video creators.
- **Claim videos** have a noticeably higher number of authors who are either **under review** or **banned**, compared to opinion videos.
- **Opinion videos** are mainly posted by active authors, with relatively few cases of banned or under-review authors.
This indicates that **claim videos** are more frequently associated with authors who are under review or have been banned.

```python
# Calculate the median view count for each author ban status
median_views = data.groupby('author_ban_status')['video_view_count'].median().reindex(['active', 'under review', 'banned'])

# Create bar plot
plt.figure(figsize=(6, 4))
sns.barplot(x=median_views.index, y=median_views.values, palette=['green', 'gold', 'orange'])

# Add labels and title
plt.title('Median Video View Count by Author Ban Status')
plt.xlabel('Author Ban Status')
plt.ylabel('Median View Count')
plt.tight_layout()
plt.show()
```

The median view counts for **non-active authors** (those under review or banned) are significantly higher than those for **active authors**. This indicates that videos from non-active authors tend to receive much more attention and visibility.

Since non-active authors are more likely to post **claim videos**, and these videos receive higher engagement, the variable `video_view_count` could serve as a useful indicator of **claim status** in a predictive model.

This suggests that claim videos are more likely to go viral or attract greater attention from viewers.

### **Task 4. Determine outliers**

```python
# List of count variables to analyze
count_columns = [
    'video_view_count',
    'video_like_count',
    'video_share_count',
    'video_download_count',
    'video_comment_count'
]

# Loop through each count variable
for column in count_columns:
    # Calculate Q1 and Q3
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    
    # Calculate IQR
    IQR = Q3 - Q1
    
    # Calculate median
    median = data[column].median()
    
    # Define outlier threshold
    outlier_threshold = median + 1.5 * IQR
    
    # Count number of values above the threshold
    outlier_count = (data[column] > outlier_threshold).sum()
    
    # Print the result
    print(f"Number of outliers, {column}: {outlier_count}")
```
